```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import layers, Model
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, Flatten, Concatenate, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import MeanAbsoluteError
import pandas as pd
import numpy as np

# Merge rating dan tempat wisata
merged_df = pd.merge(tourism_rating_df, tourism_with_id_df[['Place_Id', 'Rating', 'Price', 'Category', 'City']], on='Place_Id', how='left')

# Encode user dan place ke ID numerik
user_ids = merged_df['User_Id'].astype('category').cat.codes.values
place_ids = merged_df['Place_Id'].astype('category').cat.codes.values
ratings = merged_df['Place_Ratings'].values

n_users = len(np.unique(user_ids))
n_places = len(np.unique(place_ids))

# Model input
user_input = Input(shape=(1,))
place_input = Input(shape=(1,))

# Embedding layers
user_embedding = Embedding(input_dim=n_users, output_dim=20)(user_input)
place_embedding = Embedding(input_dim=n_places, output_dim=20)(place_input)

# Flatten
user_vec = Flatten()(user_embedding)
place_vec = Flatten()(place_embedding)

# Gabung dan Dense layers
concat = Concatenate()([user_vec, place_vec])
dense = Dense(64, activation='relu')(concat)
drop = Dropout(0.2)(dense)
dense2 = Dense(64, activation='relu')(drop)
output = Dense(1)(dense2)

model = Model(inputs=[user_input, place_input], outputs=output)
model.compile(optimizer=Adam(learning_rate=0.001), loss='mse', metrics=[MeanAbsoluteError()])

# Latih model
history = model.fit(
    [user_ids, place_ids], ratings,
    epochs=100, batch_size=32,
    validation_split=0.1,
    verbose=1
)

# Print evaluasi akhir pada data validasi
final_mae = history.history['val_mean_absolute_error'][-1]
print(f"Mean Absolute Error pada data validasi: {final_mae:.4f}")
```
